# Programming Interview
## Task 1
Combine the given lists of http and rdp observations into one data structure.
## Task 2
For the given list of IP addresses, print those that have both a rdp server that supports CredSSP and an http server that runs nginx.

```python
import json
from pprint import pprint
```

```python
folder = '/Users/collins.huff/Downloads'
http_observations = []
with open(f'{folder}/http_interview.json') as f:
    for line in f:
        http_observations.append(json.loads(line))
        
rdp_observations = []
with open(f'{folder}/rdp_interview.json') as f:
    for line in f:
        rdp_observations.append(json.loads(line))
        
ips_to_query = []
with open(f'{folder}/ips_to_query.txt') as f:
    for line in f:
        ips_to_query.append(line.strip())
       
pprint(http_observations[:1])
pprint(rdp_observations[:1])
pprint(ips_to_query)
```

## Task 1 Solution

```python
observations = {}
for http in http_observations:
    observations[http['ipv4']] = {'http': http}
for rdp in rdp_observations:
    if rdp['ipv4'] in observations:
        observations[rdp['ipv4']]['rdp'] = rdp
    else:
        observations[rdp['ipv4']] = {'rdp': rdp}
```

## Task 2 Solution

```python
for ip in ips_to_query:
    if ip in observations:
        if 'http' in observations[ip]:
            for header in observations[ip]['http']['httpHeaders']:
                if header['name'] == 'Server' and header['value'] == 'nginx':
                    if 'rdp' in observations[ip]:
                        for proto in observations[ip]['rdp']['protocols']['data']:
                            if proto['key'] == 'CredSSP' and proto['value'] == True:
                                print(ip)
```
